### 3. Preparación de Datos [0.5 puntos]

```python
# transformación de columnas a categoricas

dataset['credit_score'] = dataset['credit_score'].astype('category')

# columnas categoricas

cat_cols = ['occupation', 'payment_of_min_amount', 'payment_behaviour']
dataset[cat_cols] = dataset[cat_cols].astype('category')

# columnas numericas

num_cols = ['age', 'annual_income', 'monthly_inhand_salary', 
            'num_bank_accounts', 'num_credit_card', 'interest_rate', 
            'num_of_loan', 'delay_from_due_date', 'num_of_delayed_payment', 
            'changed_credit_limit', 'num_credit_inquiries', 'outstanding_debt', 
            'credit_utilization_ratio', 'credit_history_age', 
            'total_emi_per_month', 'amount_invested_monthly', 'monthly_balance']
```

```python
# importar column transformer

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, MinMaxScaler
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.feature_selection import SelectPercentile, f_classif
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(sparse_output=False))    
])

preprocessor = ColumnTransformer(transformers=[
    ('num', num_transformer, num_cols),
    ('cat', cat_transformer, cat_cols)],
    remainder='passthrough',
    sparse_threshold = 0)

preprocessor = preprocessor.set_output(transform="pandas")

# Pipeline

from sklearn.pipeline import Pipeline

dataset_processed = preprocessor.fit_transform(dataset.dropna())

dataset_processed.head()

```

```python
# crear holdout

from sklearn.model_selection import train_test_split

X = dataset.drop(['customer_id','credit_score'], axis=1)
y = dataset['credit_score']

# crear conjunto de entrenamiento, validacion y prueba

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=123)
```

```python
# modelos de clasificacion

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.dummy import DummyClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
import lightgbm as lgb

# Definir la métrica seleccionada (F1 Score para la clase 1)
selected_metric = 'f1_class_1'

metrics_dict = {'Model': [], selected_metric: []}

# definir modelos

models = {
    'Logistic Regression': LogisticRegression(random_state=123),
    'Random Forest': RandomForestClassifier(random_state=123),
    'SVC': SVC(random_state=123),
    'KNN': KNeighborsClassifier(),
    'Dummy Classifier': DummyClassifier(strategy='stratified', random_state=123),
    'Decision Tree': DecisionTreeClassifier(random_state=123),
    'XGB': XGBClassifier(),
    'LGBM': lgb.LGBMClassifier(verbose=0, random_state=123)
}

# Iterar sobre modelos
for model_name, model in models.items():
    # Crear la pipeline para el modelo actual
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', model)
    ])

    # Ajustar la pipeline al conjunto de datos
    pipeline.fit(X_train, y_train)

    # Realizar predicciones
    y_pred = pipeline.predict(X_test)

    # Calcular la métrica seleccionada
    f1_class_1 = f1_score(y_test, y_pred, pos_label=1)

    # Imprimir classification_report
    print(f"Classification Report for {model_name}:\n", classification_report(y_test, y_pred, zero_division=1))

    # Guardar la métrica en el diccionario
    metrics_dict['Model'].append(model_name)
    metrics_dict[selected_metric].append(f1_class_1)

# Convertir el diccionario de métricas a un DataFrame
metrics_df = pd.DataFrame(metrics_dict)

# Ordenar el DataFrame por el valor de la métrica de mayor a menor
metrics_df = metrics_df.sort_values(by=selected_metric, ascending=False)

# Imprimir el DataFrame ordenado
print("\nMetrics DataFrame:")
metrics_df
```
