### 3. Preparación de Datos 

En el punto anterior, se pudo apreciar que existen `outliers` en ciertas variables, los cuales además de entorpecer el estudio, no tienen sentido alguno, como edades fuera del orden de dos dígitos o números negativos de cuentas bancarias. Aquellos valores se eliminarán antes del preprocesamiento.

```python
dataset_2 = dataset.copy()

# age
dataset_2 = dataset_2[(dataset_2['age'] > 0) & (dataset_2['age'] < 120)]

# num_bank_accounts
dataset_2 = dataset_2[(dataset_2['num_bank_accounts'] > 0) & (dataset_2['num_bank_accounts'] < 20)]

# num_credit_card
dataset_2 = dataset_2[dataset_2['num_credit_card'] < 20]

# num_of_loan
dataset_2 = dataset_2[(dataset_2['num_of_loan'] > 0) & (dataset_2['num_of_loan'] < 50)]
```

```python
# graficos de distribucion

import matplotlib.pyplot as plt

dataset_2.hist(bins=50, figsize=(20,15))
plt.show()
```

Luego, para la preparación de los datos, se creará una `pipeline` que tenga como primer paso la transformación de las columnas a un formato adecuado, es decir, escalar las variables numéricas como corresponda y transformar las categorías en columnas binarias. Para ello se comienza definiendo qué columnas son numéricas y categóricas.

```python
# transformación de columnas a categoricas

dataset_2['credit_score'] = dataset_2['credit_score'].astype('category')

# columnas categoricas

cat_cols = ['occupation', 'payment_of_min_amount', 'payment_behaviour']
dataset_2[cat_cols] = dataset_2[cat_cols].astype('category')

# columnas numericas

num_cols = ['age', 'annual_income', 'monthly_inhand_salary', 
            'num_bank_accounts', 'num_credit_card', 'interest_rate', 
            'num_of_loan', 'delay_from_due_date', 'num_of_delayed_payment', 
            'changed_credit_limit', 'num_credit_inquiries', 'outstanding_debt', 
            'credit_utilization_ratio', 'credit_history_age', 
            'total_emi_per_month', 'amount_invested_monthly', 'monthly_balance']
```

#### 3.1 Preprocesamiento con `ColumnTransformer`

Se crea la `pipeline` para el preprocesamiento considerando un caso para las variables numéricas y otro para las categorías. Para las variables numéricas se hace uso del método `StandardScaler`, mientras que para las categorías se utiliza `OneHotEncoder`.

```python
# importar column transformer

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, MinMaxScaler
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.feature_selection import SelectPercentile, f_classif
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(sparse_output=False))    
])

preprocessor = ColumnTransformer(transformers=[
    ('num', num_transformer, num_cols),
    ('cat', cat_transformer, cat_cols)],
    remainder='passthrough',
    sparse_threshold = 0)

preprocessor = preprocessor.set_output(transform="pandas")

# Pipeline

from sklearn.pipeline import Pipeline

dataset_processed = preprocessor.fit_transform(dataset_2)

dataset_processed.head()

```

#### 3.2 Holdout 

Se crean los conjuntos de entrenamiento, validación y de prueba.

```python
# crear holdout

from sklearn.model_selection import train_test_split

X = dataset_2.drop(['customer_id','credit_score'], axis=1)
y = dataset_2['credit_score']

# crear conjunto de entrenamiento, validacion y prueba

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=123)
```

#### 3.3 Datos nulos.

Como se pudo ver al inicio de la sección de preprocesamiento, se decidió filtrar observaciones con valores atípicos, lo cual eliminó algunos de los valores nulos presentes en el conjunto de datos. Para los demás datos sin valores, se decidió imputar utilizando el promedio para las variables numéricas y utilizando la moda para variables categóricas, de esta forma se evita quitar más observaciones de las que ya se descartaron anteriormente.

### 4. Baseline

Se crea una nueva `pipeline` que contempla el preprocesamiento anterior y se le agrega un modelo de clasificación, el cual servirá como `baseline` para futuras optimizaciones de hiperparámetros. Se utilizarán los siguientes modelos de clasificación:

- `Dummy` con estrategia estratificada.
- `LogisticRegression`.
- `KNeighborsClassifier`.
- `DecisionTreeClassifier`
- `SVC`
- `RandomForestClassifier` 
- `LightGBMClassifier` (del paquete `lightgbm`)
- `XGBClassifier` (del paquete `xgboost`).

A continuación, se mostrarán los informes de clasificación para cada modelo y un resumen con el valor de `F1 score` para la clase positiva: 

```python
# modelos de clasificacion

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.dummy import DummyClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
import lightgbm as lgb
import time


# Definir la métrica seleccionada (F1 Score para la clase 1)
selected_metric = 'f1_class_1'

metrics_dict = {'Model': [], selected_metric: [], 'f1_class_1_val': [], 'f1_class_1_prom': []}

# definir modelos

models = {
    'Logistic Regression': LogisticRegression(random_state=123),
    'Random Forest': RandomForestClassifier(random_state=123),
    'SVC': SVC(random_state=123),
    'KNN': KNeighborsClassifier(),
    'Dummy Classifier': DummyClassifier(strategy='stratified', random_state=123),
    'Decision Tree': DecisionTreeClassifier(random_state=123),
    'XGB': XGBClassifier(),
    'LGBM': lgb.LGBMClassifier(verbose=0, random_state=123)
}

# Iterar sobre modelos
for model_name, model in models.items():
    # Crear la pipeline para el modelo actual
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', model)
    ])

    # Iniciar el temporizador
    start_time = time.time()

    # Ajustar la pipeline al conjunto de entrenamiento
    pipeline.fit(X_train, y_train)

    # Detener el temporizador y calcular el tiempo de ejecución
    elapsed_time = time.time() - start_time

    # Realizar predicciones en el conjunto de prueba
    y_pred_test = pipeline.predict(X_test)

    # Calcular la métrica seleccionada en el conjunto de prueba
    f1_class_1_test = f1_score(y_test, y_pred_test, pos_label=1)

    # Realizar predicciones en el conjunto de validación
    y_pred_val = pipeline.predict(X_val)

    # Calcular la métrica seleccionada en el conjunto de validación
    f1_class_1_val = f1_score(y_val, y_pred_val, pos_label=1)

    # Imprimir classification_report, tiempo de ejecución y métrica en conjunto de validación
    print(f"Classification Report for {model_name} (Test Set):\n", classification_report(y_test, y_pred_test, zero_division=1))
    print(f"Classification Report for {model_name} (Validation Set):\n", classification_report(y_val, y_pred_val, zero_division=1))
    print(f"Elapsed Time for {model_name}: {elapsed_time} seconds")
    print(f"F1 Score for {model_name} (Test Set): {f1_class_1_test}")
    print(f"F1 Score for {model_name} (Validation Set): {f1_class_1_val}\n")

    # Guardar las métricas en el diccionario
    metrics_dict['Model'].append(model_name)
    metrics_dict[selected_metric].append(f1_class_1_test)
    metrics_dict['f1_class_1_val'].append(f1_class_1_val)
    metrics_dict['f1_class_1_prom'].append((f1_class_1_test + f1_class_1_val)/2)

# Convertir el diccionario de métricas a un DataFrame
metrics_df = pd.DataFrame(metrics_dict)

# Ordenar el DataFrame por el valor de la métrica de mayor a menor
metrics_df = metrics_df.sort_values(by=selected_metric, ascending=False)

# Imprimir el DataFrame ordenado
print("\nMetrics DataFrame:")
metrics_df
```

Es posible ver que todos los modelos en su `baseline` clasifican mejor que el `Dummy` para la métrica escogida, destacando el rendimiento de `LGBM` y `Random Forest`, que llegan a 0.60 pts al promediar entre el conjunto de prueba y validación. Es probable que el rendimiento de estos modelos haya sido superior al resto debido a su construcción, siendo ambos algoritmos basados en el ensamble de árboles, combinando sus predicciones para mejorar el rendimiento general del modelo. 

Con respecto al tiempo de entrenamiento, el modelo que toma más tiempo en entrenarse es `Random Forest` con 3 segundos, mientras que `LGBM` solo toma 0.2 segundos. Debido a esto, y a los valores de `baseline`, el modelo más adecuado a optimizar sus hiperparámetros es `LGBM`. 

### 5. Optimización del Modelo

En esta sección se toman como modelos a optimizar `LGBM` y `Random Forest`, dónde se actualizarán sus hiperparámetros más relevante utilizando el framework `optuna`. Se realizarán 100 iteraciones para encontrar la configuración que mejore los valores para ambos modelos.

Para ello, se definen dos funciones objetivo, una por cada modelo a optimizar con los hiperparámetros de interés y los valores en los que se buscará el mejor rendimiento de `F1 score`. Al igual que en las secciones anteriores, los modelos se trabajarán a través de una pipeline, la cual esta vez posee un término extra que realiza selección de características según el percentil de aporte al `F1 score`.


```python
# optimizacion de hiperparametros

import optuna
from sklearn.metrics import make_scorer
from sklearn.feature_selection import SelectPercentile, f_classif

# metrica a optimizar   

selected_metric = 'f1_class_1'

# crear una funcion que retorne f1 para la clase 1

def f1_class_1(y_true, y_pred):
    return f1_score(y_true, y_pred, pos_label=1)

scorer = make_scorer(f1_class_1)

# crear funcion para optimizar lgbm

def objective_lgbm(trial):
    # Crear un diccionario con los parámetros
    params = params = {'num_leaves': trial.suggest_int('num_leaves', 20, 150),
                       'learning_rate': trial.suggest_float('learning_rate', 0.001, 0.05),
                       'max_depth': trial.suggest_int('max_depth', 2, 256),
                       'features__percentile': trial.suggest_int('features__percentile', 20, 100),
                       'min_child_samples': trial.suggest_int('min_child_samples', 2, 256),
                       'min_child_weight': trial.suggest_int('min_child_weight', 0.1, 0.5),
                       'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
                       'percentile': trial.suggest_int('percentile', 20, 100),
                       }

    # Crear la pipeline con los parámetros
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('features', SelectPercentile(f_classif, percentile=params['percentile'])),
        ('classifier', lgb.LGBMClassifier(**params, verbose=-1))
    ])

    pipeline.fit(X_train, y_train)

    # Calcular el score en el conjunto de prueba
    score_test = f1_class_1(y_test, pipeline.predict(X_test))

    # Calcular el score en el conjunto de validación
    score_val = f1_class_1(y_val, pipeline.predict(X_val))

    # Tomar la media de los scores de prueba y validación
    final_score = (score_test + score_val) / 2

    return final_score

# crear funcion para optimizar random forest

def objective_rf(trial):
    # Crear un diccionario con los parámetros
    params = {'bootstrap': True,
              'ccp_alpha': trial.suggest_float('ccp_alpha', 0.0, 0.5),
              'class_weight': trial.suggest_categorical('class_weight', ['balanced', 'balanced_subsample', None]),
              'criterion': trial.suggest_categorical('criterion', ['gini', 'entropy', 'log_loss']),
              'max_depth': trial.suggest_int('max_depth', 2, 256),
              'max_leaf_nodes': trial.suggest_int('max_leaf_nodes', 2, 256),
              'min_impurity_decrease': trial.suggest_float('min_impurity_decrease', 0.0, 0.5),
              'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 10),
              'min_samples_split': trial.suggest_int('min_samples_split', 2, 10),
              'min_weight_fraction_leaf': trial.suggest_float('min_weight_fraction_leaf', 0.0, 0.5),
              'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
              }

    features_params = {
        'percentile': trial.suggest_int('percentile', 20, 100)
    }

    # Crear la pipeline con los parámetros
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('features', SelectPercentile(f_classif, **features_params)),
        ('classifier', RandomForestClassifier(**params, random_state=123))
    ])
    
    pipeline.fit(X_train, y_train)

    # Calcular el score en el conjunto de prueba

    score_test = f1_class_1(y_test, pipeline.predict(X_test))

    # Calcular el score en el conjunto de validación

    score_val = f1_class_1(y_val, pipeline.predict(X_val))

    # Tomar la media de los scores de prueba y validación

    final_score = (score_test + score_val) / 2

    return final_score
```

```python
# optimizar lgbm

import warnings

optuna.logging.set_verbosity(optuna.logging.WARNING)

#warnings.filterwarnings("ignore", category=lgb.LGBMWarning)

study_lgbm = optuna.create_study(direction='maximize')
study_lgbm.optimize(objective_lgbm, n_trials=100, show_progress_bar=True)
```

Al terminar la optimización de `LGBM`, es posible ver que el mejor valor de `F1 score` se encuentra en la iteración 42, obteniendo un valor de 0.617 pts, mejorando el valor promedio pre-optimización.

Por otro lado, con `Random Forest` se obtiene un `F1 score` de 0.649 pts en la iteración 84, mejorando notablemente el valor de la `baseline`.

Al graficar, es posible apreciar como los valores se incrementan considerablemente desde la décima iteración, los cuales se estabilizan. Mostrando que es posible entrenar al modelo en menos iteraciones y obtener el resultado en menor tiempo.
