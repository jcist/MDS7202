5. ¿Que comandos de `Git` sirven para crear un repositorio y hacer seguimiento a cambios en los archivos?

<code>Escriba sus respuestas acá:</code>

1. Git es un DVCS (Distributed Version Control System), es decir, es un sistema que registra los cambios realizados en un archivo o conjunto de archivos a lo largo del tiempo, lo cual nos permite llevar un registro de los cambios en un proyecto y trabajar en ramas paralelas, lo que facilita el trabajo en equipo (VCS). Al ser además distribuido, los clientes pueden acceder a una replica del repositorio completo, no sólo la última copia instantánea de archivos; tienen la posibilidad de clonar este repositorio en completitud, lo que permite establecer varios flujos de trabajo que no son posibles en sistemas centralizados.

2. Diferencia 1: Ubicación
    Git se instala en el sistema local, mientras que GitHub está alojada en la web.

   Diferencia 2: Funcionamiento
    Git es un software que funciona como herramienta de lineas de comandos, GitHub es un servicio que entrega una interfaz gráfica de trabajo.
    
   Diferencia 3: ¿Qué son?
    Git es una herramienta para gestionar diferentes versiones de ediciones realizadas en archivos de un repositorio Git, mientras que GitHub es un espacio para subir una copia del repositorio Git.

3. Un hash es el resultado de una operación criptográfica a partir del contendo del objeto a identificar (hash) que genera identificadores de 40 caractéres (letras y números) únicos e irrepetibles, en el caso de Git, se utiliza el hash SHA1 y puede apuntar a un blob, un commit o un árbol. El hash de un commit se puede encontrar en su información junto con el nombre del autor, fecha y descripción, un ejemplo de hash de commit es el siguiente identificador "3e889ee4e7eeacfa7cf667a693344c31068746bf"

4. Las unidades básicas de almacenamiento de Git son blob, commit y árbol. Todo el contenido de un archivo se almacena en un blob, cuando el archivo es modificado, un nuevo blob almacenará el archivo completo con todos los nuevos cambios. Los archivos se almacenan en algún directorio o carpeta y estas carpetas pueden contener más carpetas/directorios. Un árbol representa directorios para blobs y otros árboles, por lo que siempre hay un árbol en la raíz, apuntando al árbol que contiene cosas. Además, Git se compone de 3 estados: el directorio de trabajo (Working Directory) que es donde se almacenan los archivos del proyecto, el área de preparación (Staging Area) que es donde se preparan los archivos para ser enviados al repositorio y finalmente el git directory (Repositoy) que es donde se almacenan todos los cambios y versiones del proyecto a lo largo del tiempo.

5. Para crear un repositorio usamos el comando

*   git init nombre-del-repo: Inicia un nuevo repositorio en la carpeta actual.

Y para hacer seguimiento de los cambios usamos el comando: 
   
*   git status: Muestra el estado actual del repositorio.
*   git log: Muestra el historial de cambios realizados.

Otros comandos útiles son: 
*   git add: Añade cambios en los archivos al área de preparación.
*   git commit: Guarda los cambios realizados en los archivos en el repositorio.
*   git branch: Para ver las ramas disponibles en el repositorio o crea una nueva rama.
*   git checkout: Cambia entre ramas o restaura archivos a versiones anteriores.
*   git merge: Fusiona cambios entre dos ramas del repositorio.
*   git remote: Muestra o configura conexiones con repositorios remotos.
*   git pull: Obtener el registro de cambios de un repositorio remoto.
*   git push: Envía los cambios realizados en el repositorio local a un repositorio remoto

Por último, investigue y comente para qué sirven los siguientes comandos:

11. `git stash` sirve para guardar temporalmente los cambios que se han realizado en el repositorio, para luego poder recuperarlos en otro momento.
12. `git reset` sirve para deshacer cambios en el repositorio, ya sea para deshacer un commit o para deshacer un git add. La diferencia con `git revert` es que `git reset` deshace los cambios de manera permanente, mientras que `git revert` deshace los cambios de manera temporal, creando un nuevo commit que deshace los cambios realizados.
