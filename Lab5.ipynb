### 1.3.4 Informaci칩n por Categor칤as [1 punto]

````python
# promedio de notas GPA
df_grades['GPA'] = (df_grades['math score'] + df_grades['reading score'] + df_grades['writing score']) / 3
df_grades['GPA'] = df_grades['GPA'].round(2)

# Definir un diccionario para mapear niveles educacionales
level_simplificated = {'some high school': 'school', 'some college': 'school', 'high school': 'school', "bachelor's degree": 'college', 
                       "associate's degree": 'college', "master's degree": 'postgraduate'}

# Aplicar el diccionario de mapeo a la columna 'parental level of education'
df_grades['simple parental level of education'] = df_grades['parental level of education'].map(level_simplificated)

df_grades.head()
````

````python
# calcular la media y redondear a dos decimales
def mean_and_round(series):
    return round(series.mean(), 2)

# doble groupby
df_grades_doble_groupby = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).agg(
    {
        'math score': mean_and_round,
        'reading score': mean_and_round,
        'writing score': mean_and_round,
        'GPA': mean_and_round
    }
).reset_index()

# conteo
df_grades_doble_groupby['count'] = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).size().reset_index(name='count')['count']

# porcentajes
df_grades_doble_groupby['percentage'] = ((df_grades_doble_groupby['count'] / df_grades_doble_groupby['count'].sum())*100).round(2)
df_grades_doble_groupby['percentage'] = df_grades_doble_groupby['percentage'].astype(str) + '%'

df_grades_doble_groupby
````

````python
other_grades = pd.read_csv('other_grades.csv')

other_grades['science score'] = round(other_grades['science score']*0.06 + 1, 1)

other_grades
````

````python
# Realizar un outer join en la columna 'names'
outer_join_df = pd.merge(other_grades, df_grades, on='names', how='outer')
print('\nOuter Join:')
outer_join_df

# obtener el numero de filas y columnas
print('Numero de filas y columnas:', outer_join_df.shape)
````

````python
# Realizar un left join en la columna 'names'
left_join_df = pd.merge(other_grades, df_grades, on='names', how='left')
print('\nLeft Join:')
left_join_df

# obtener el numero de filas y columnas
print('Numero de filas y columnas:', left_join_df.shape)
````

````python
# Realizar un right join en la columna 'names'
right_join_df = pd.merge(other_grades, df_grades, on='names', how='right')
print('\nRight Join:')
right_join_df

# obtener el numero de filas y columnas
print('Numero de filas y columnas:', right_join_df.shape)
````

````python
# Realizar un inner join en la columna 'names'
inner_join_df = pd.merge(other_grades, df_grades, on='names', how='inner')
print('Inner Join:')
inner_join_df

# obtener el numero de filas y columnas
print('Numero de filas y columnas:', inner_join_df.shape)
````

> **Respuesta**: 

Al utilizar outer join, se unen los datos que est치n disponibles en los dataframes, sin importar si las llaves primarias se encuentran en ambas bases y completando con NaN los atributos faltantes en caso de que una instancia se encuentre en una de las dos bases. En este caso, se obtienen 1000 observaciones, lo cual coincide con other_grades. Esto pues las llaves del dataset original son subconjunto de other_grades.

Al utilizar left join, se unen todos los datos de la tabla de la izquierda (en este caso, other_grades) a partir de su llave primaria con todas aquellas que coincidan en la tabla de la derecha. En este caso, como df_grades es subconjunto de other_grades, todas las observaciones de la nueva base corresponden a alumnos que est치n en other_grades. Por lo tanto, se obtienen 1000 filas y coincide con el outer join

Al utilizar right join, se unen todos los datos de la tabla de la derecha (other_grades) a partir de su llave primaria con todas aquellas que coincidan en la tabla de la izquierda (other_grades). En este caso, como df_grades es subconjunto de other_grades, todas las instancias de la nueva base corresponden a alumnos que est치n en df_grades. Por lo tanto, se obtienen 875 filas sin necesidad de completar atributos con datos nulos.

Al utilizar inner join, se unen los datos que est치n disponibles en ambos dataframes a partir de su llave primaria, de este modo todos los atributos muestran total completitud de los datos. En este caso, se obtienen 875 instancias, lo cual coincide con el tama침o del dataset m치s peque침o (df_grades) y con el resultado del right join.
